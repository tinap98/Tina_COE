# Assignment on Store_sales dataset

**Load the Dataset**

```python
import pandas as pd
store_sales = pd.read_csv('store_sales.csv')
store_sales
```

**Calculate the total sales for each store across all months.**

```python
store_sales['total_sales'] = store_sales.iloc[:, 2:].sum(axis=1)
store_sales[['store_id', 'total_sales']]
```

**Find the average sales for each month across all stores**

```python
avg_sales = store_sales.iloc[:, 2:14].mean()
avg_sales
```

**Identify the store with the highest total sales.**

```python
store_sales['total_sales'] = store_sales.iloc[:, 2:].sum(axis=1)
highest_sales = store_sales.loc[store_sales['total_sales'].idxmax()]
highest_sales[['store_id', 'total_sales']]
```

**Calculate the total sales for each city.**

```python
store_sales['total_sales'] = store_sales.iloc[:, 2:].sum(axis=1)
city_sales = store_sales.groupby('city')['total_sales'].sum()
city_sales
```

**List stores with total sales greater than 200**

```python
stores = store_sales[store_sales['total_sales'] > 200]
stores
```
