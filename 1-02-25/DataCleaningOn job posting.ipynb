```python
import re

def clean_text(text):
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['JobDescription'] = df['JobDescription'].apply(lambda x: clean_text(str(x)))
df['JobRequirment'] = df['JobRequirment'].apply(lambda x: clean_text(str(x)))
df['RequiredQual'] = df['RequiredQual'].apply(lambda x: clean_text(str(x)))
```

```python
from nltk.tokenize import WordPunctTokenizer
import re

tokenizer = WordPunctTokenizer()

def clean_tokens(tokens):
    return [token.lower() for token in tokens if re.match(r'\w+', token)]  # Only keep alphanumeric words

df['JobDescription'] = df['JobDescription'].apply(lambda x: clean_tokens(tokenizer.tokenize(str(x).lower())))
df['JobRequirment'] = df['JobRequirment'].apply(lambda x: clean_tokens(tokenizer.tokenize(str(x).lower())))

df[['JobDescription','JobRequirment']].head()
```

```python
import re

def normalize_title(title):
    title = title.lower()

    title = re.sub(r'[^\w\s]', '', title)

    title = re.sub(r'\s+', ' ', title).strip()


    return title

df['NormalizedTitle'] = df['JobDescription'].apply(lambda x: normalize_title(str(x)))
df[['NormalizedTitle']].head()
```

```python
import re

def standardize_terms(text, term_mapping):
    for term, standard in term_mapping.items():
        text = re.sub(r'\b' + re.escape(term) + r'\b', standard, text, flags=re.IGNORECASE)
    return text

term_mapping = {
    'developer': 'software engineer',
    'designer': 'UI/UX designer',
    'manager': 'project manager',
    'lead': 'team lead',
    'consultant': 'business consultant',
}

df['StandardizedJobDescription'] = df['JobDescription'].apply(lambda x: standardize_terms(str(x), term_mapping))
df['StandardizedJobRequirment'] = df['JobRequirment'].apply(lambda x: standardize_terms(str(x), term_mapping))

df[['StandardizedJobDescription','StandardizedJobRequirment']].head()
```

```python
import re

skills_list = [
    'python', 'java', 'javascript', 'sql', 'html', 'css', 'angular', 'react', 'aws', 
    'machine learning', 'data science', 'deep learning', 'android', 'ios', 'nodejs', 
    'devops', 'docker', 'kubernetes', 'cloud computing', 'cybersecurity', 'sql'
]

def extract_skills(text, skills_list):
    tokens = re.findall(r'\b\w+\b', text.lower())
    
    found_skills = [skill for skill in skills_list if skill in tokens]
    
    return found_skills

df['ExtractedJobDescriptionSkills'] = df['JobDescription'].apply(lambda x: extract_skills(str(x), skills_list))
df['ExtractedJobRequirmentSkills'] = df['JobRequirment'].apply(lambda x: extract_skills(str(x), skills_list))

df[['ExtractedJobDescriptionSkills', 'ExtractedJobRequirmentSkills']].head()
```

```python
df['Deadline'].fillna(df['OpeningDate'], inplace=True)

df['Jobstartdelay'] = (df['StartDate'] - df['OpeningDate']).dt.days
df['Dayslefttoapply'] = (df['Deadline'] - df['OpeningDate']).dt.days

df.loc[df['Dayslefttoapply'] < 0, 'Dayslefttoapply'] = 0
df.loc[df['Jobstartdelay'] < 0, 'Jobstartdelay'] = 0

df['Year_Opening'] = df['OpeningDate'].dt.year
df['Month_Opening'] = df['OpeningDate'].dt.month
df['Year_Deadline'] = df['Deadline'].dt.year
df['Month_Deadline'] = df['Deadline'].dt.month
df['Year_Start'] = df['StartDate'].dt.year
df['Month_Start'] = df['StartDate'].dt.month

df[['StartDate', 'OpeningDate', 'Deadline', 'Dayslefttoapply', 'Jobstartdelay', 'Year_Opening', 'Month_Opening', 'Year_Deadline', 'Month_Deadline', 'Year_Start', 'Month_Start']]
```

```python
df[['City', 'Country']] = df['Location'].str.split(',', expand=True, n=1)

df['City'] = df['City'].str.strip()
df['Country'] = df['Country'].str.strip()

df['Country'] = df['Country'].fillna('Unknown')
```

```python
import re

def extract_salary_range(salary):
    match = re.search(r"(\d+)[Kk]?-(\d+)[Kk]?", salary)
    if match:
        return int(match.group(1)) * 1000, int(match.group(2)) * 1000
    return None, None

df[['min_salary', 'max_salary']] = df['Salary'].apply(lambda x: pd.Series(extract_salary_range(x)))
```

```python
df['IT'] = df['IT'].astype(str)

df['IT_related_job'] = df['IT'].str.contains("IT|Software|Developer|Engineer", case=False, na=False)
```

```python

education_map = {
    "Bachelors": "Undergraduate",
    "Masters": "Postgraduate",
    "PhD": "Doctorate",
    "High School": "High School"
}

df['Education_Level'] = df['Eligibility'].replace(education_map)

def extract_experience(audience):
    match = re.search(r"(\d+)-(\d+)\s*years?", audience)
    if match:
        return f"{match.group(1)}-{match.group(2)} years"
    elif "entry-level" in audience.lower():
        return "0-1 years"
    elif "senior" in audience.lower():
        return "5+ years"
    return "Unknown"

df['Experience_Range'] = df['Audience'].apply(extract_experience)
```
