```python
import pandas as pd
uber_data_csv = pd.read_csv("Uber.csv", nrows=100)
```

```python
print("First 5 rows:")
uber_data_csv.head()
```

```python
print("\nLast 2 columns:")
uber_data_csv.iloc[:, -2:]
```

```python
print("\nRows 50 to 100:")
uber_data_csv.iloc[50:100]
```

```python
print("\nAll rows from first 3 columns:")
uber_data_csv.iloc[:, :3]
```

```python
uber_data_csv['START_DATE*'] = pd.to_datetime(uber_data_csv['START_DATE*'], errors='coerce')
uber_data_csv['END_DATE*'] = pd.to_datetime(uber_data_csv['END_DATE*'], errors='coerce')
```

```python
uber_data_csv['MILES*'] = pd.to_numeric(uber_data_csv['MILES*'], errors='coerce')
```
