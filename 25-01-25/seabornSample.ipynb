```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.histplot(data=tips,x='total_bill',kde=True,bins=20,color='skyblue')
plt.title("distribution of total bills")
plt.show()
```

```python
sns.kdeplot(data=tips,x='total_bill',fill=True,color='purple')
plt.title("KDE plot of total bills")
plt.show()
```

```python
sns.barplot(data=tips,x="day",y="total_bill",palette="cubehelix",hue="day")
plt.title("Average total bill by day")
plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",palette="YlGn")
plt.title("Box plot of total bills by day and gender")
plt.show()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time",palette="crest")
plt.title("scatter plot of total bill vs tip")
plt.show()
```

```python
sns.lineplot(data=tips,x="size",y="tip",marker="o",color="skyblue")
plt.title("line plot of party size vs tip")
plt.show()
```

```python
sns.pairplot(data=tips,hue="sex",palette="husl")
plt.show()
```

```python

numeric_cols = tips.select_dtypes(include=['number']).columns
corr_matrix = tips[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
