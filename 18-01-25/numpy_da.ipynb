```python
import pandas as pd
import numpy as np
uber_data = pd.read_csv("Uber.csv") 
```

```python
a=uber_data.sort_values(by='MILES*')
a.reset_index(inplace=True,drop=True)
a
```

```python
uber_data["MILES_CAT"]=np.where(uber_data['MILES*']>100,"Long trip","Short trip")
uber_data
```

```python
a=uber_data.sort_values(by=['START*','MILES*'],ascending=[True,False])
a
```

```python
uber_data["trip_type"] = np.where(uber_data['MILES*'] <= 50, "short trip",  np.where(uber_data['MILES*'] <= 100, "medium trip", "long trip"))
uber_data
```

```python
uber_data.drop(columns='MILES_CAT')
```

```python
trip_type_counts = uber_data['trip_type'].value_counts()
print(trip_type_counts)
```
