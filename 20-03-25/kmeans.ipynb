```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
df=pd.read_csv('kmeans_dataset_with_outliers.csv')
```

```python
df
```

```python
df=df.drop('Label',axis=1)
```

```python
df
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(df)

# Step 3: Get the cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Step 4: Visualize the clusters
plt.scatter(df['X'], df['Y'], c=labels, cmap='viridis', s=100)  # Color the points by cluster
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centers')  # Plot centers
plt.title('K-Means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Define a range of cluster numbers to test
cluster_range = range(1, 11)  # Test from 1 to 10 clusters
inertia = []

# Step 2: Calculate K-Means for each number of clusters
for k in cluster_range:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    inertia.append(kmeans.inertia_)

# Step 3: Plot the elbow method
plt.plot(cluster_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.xticks(cluster_range)
plt.show()
```
