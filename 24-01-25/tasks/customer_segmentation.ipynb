```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('credit_card_customer_data.csv')
df
```

```python
# 1. Avg Credit Limit(Histogram)
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['Avg_Credit_Limit'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of Average Credit Limit')
plt.xlabel('Avg Credit Limit')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2. Spending Patterns (Scatter Plot)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Avg_Credit_Limit'], df['Total_visits_online'], alpha=0.5, color='purple')
plt.title('Spending Patterns: Avg Credit Limit vs Total Visits Online')
plt.xlabel('Avg Credit Limit')
plt.ylabel('Total Visits Online')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 3. Spending Patterns (Scatter Plot)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Avg_Credit_Limit'], df['Total_calls_made'], alpha=0.5, color='blue')
plt.title('Spending Patterns: Avg Credit Limit vs Total Calls Made')
plt.xlabel('Avg Credit Limit')
plt.ylabel('Total Calls Made')
plt.grid(True)
plt.tight_layout()
plt.show()
```
