```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
subject_columns = ['math score', 'reading score', 'writing score']
```

```python
subject_averages = df[subject_columns].mean()
```

```python
# bar chart
plt.figure(figsize=(10,6))
subject_averages.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Marks Across Subjects')
plt.xlabel('Subject')
plt.ylabel('Average Marks')
plt.xticks(rotation=0)
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
# 2 (Box Plot)
plt.figure(figsize=(10,6))
sns.boxplot(data=df[subject_columns], palette="Set3")
plt.title('Box Plot of Marks Across Subjects')
plt.xlabel('Subject')
plt.ylabel('Marks')
plt.tight_layout()
plt.show()
```
