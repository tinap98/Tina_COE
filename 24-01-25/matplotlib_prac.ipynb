```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.title("Simple line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.plot(x,y,color='red',linestyle='dashed',linewidth=2,marker='o')
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[200,300,500,100,400,200]
plt.plot(x,y)
plt.plot(x,y,color='green',linestyle='dashed',linewidth=4,marker='o')
plt.title("savings growth")
plt.xlabel("Month")
plt.ylabel("Savings amount")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color='skyblue',edgecolor='black',width=0.5)
plt.title("Basic bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
age=[18,32,23,34,21,43,56,66,43,12,22,24,43,98,69]
b=[20,30,40,50,60,70]
plt.hist(age,bins=b,color='skyblue',edgecolor='black')
plt.title("Age Distribution of customers")
plt.xlabel("Age ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python

languages = ['Python', 'Ruby', 'C++', 'Java']
usage = [40, 20, 15, 25]  
colors = ['pink', 'skyblue', 'yellow', 'lightgreen']

plt.pie(usage, labels=languages, colors=colors, autopct='%1.1f%%',startangle=160)
plt.legend()
plt.title('Programming Languages Usage')

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.scatter(x,y,color='green',s=100,edgecolor='black')
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```
